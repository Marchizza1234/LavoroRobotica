```python
# Per controllare che il numero assoluto sia pari o dispari ---> abs(numero%=0)
numeri = [0] * (3)

print("Dammi 3 numeri da controllare")
for i in range(0, 2 + 1, 1):
    numeri[i] = int(input())
for i in range(0, 2 + 1, 1):
    if numeri[i] > 0:
        print("Il numero è maggiore di 0")
    else:
        if numeri[i] < 0:
            print("Il numero è minore di 0")
        else:
            print("Il numero è 0")
    print(numeri[i])
    if fabs(numeri[i]) % 2 == 0:
        print("Il numero è pari")
    else:
        print("Il numero è dispari")
```
